# Generators

<ul><li> Are Functions that <b>'return'</b> traversable objects.</li>
    <li>Produce items <b>'one at a time'</b> when required.</li>
    <li>Are run with <b>'For'</b> loops</li></ul>
    
    
## Advantages
<ul>
    <li>Easy to implement <b>(implement_iter_()._next()_)</b> automatically</li>
    <li>Better memory management and utilization.</li>
    <li>Can be used for infinite items.</li>
    <li>Can also be used for number of iterations.</li>

</ul>

# Diif. Btwn Normal  v/s Generators functions

<table>
    <thead>
        <tr>
            <td>Generators functions</td>            
            <td>Normal functions</td>
        </tr>
    </thead>

        <tr>
            <td>Make use of 'yield' keyword</td>
            <td>Make use of 'return' keyword</td>
        </tr>
        <tr>
            <td>Run when 'next()' method call</td>
            <td>Run when name of method is called</td>
        </tr>
        <tr>
            <td>Produce item when at a time and only when required</td>
            <td>Produce item at once.</td>
        </tr>
    </tbody>
</table>

```python
def func(a):
    yield a*a
    
    
s = func(5)
s
```

```python
next(s)
```

```python
def new(dict):
    for x,y in dict.items():
        yield x,y
        
a = {1:'ankit',2:'Garg'}
s = new(a)
print(s)
```

```python
next(s)
```

```python
next(s)
```

# Fabonacci Series

```python
def fib():
    f,s=0,1
    while True:
        yield f
        f,s=s,s+f
    
for i in fib():
    if i >50:
        break
    print(i, end=' ')
    
    
```

```python
# Using simple functions

import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb

def s(flip=2):
    x = np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
        
sb.set()
s()
plt.show()
    
    
```

```python
# Using generators

import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb

def s(flip=2):
    x = np.linspace(0,14,100)
    for i in range(1,5):
        yield(plt.plot(x,np.sin(x+i*.5)*(7-i)*flip))
        
sb.set()
s=s()
plt.show()
    
print(next(s)) 
print(next(s)) 
```
